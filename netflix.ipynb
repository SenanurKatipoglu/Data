```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb

desktop_path = "~/Desktop" 
file_path = f"{desktop_path}/netflix.csv"

data = pd.read_csv(file_path)

data
```

```python
data.head()
```

```python
data.info()
```

```python
data_mean = data['imdb_score'].mean()
```

```python
print("dataset imdb score mean =", data_mean)
```

```python
above_mean = data[data['imdb_score']>= data_mean]
below_mean = data[data['imdb_score']< data_mean]

plt.figure(figsize=(10, 6))

sb.scatterplot(x='type', y='imdb_score', data=above_mean, color='green', label='Above of Mean')

sb.scatterplot(x='type', y='imdb_score', data=below_mean, color='red', label='Below of Mean')

plt.axhline(data_mean, color='blue', linestyle='--', label='Mean IMDb')


plt.title('IMDb Score and Mean')
plt.xlabel('type')
plt.ylabel('imdb_score')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()


```
